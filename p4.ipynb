```python
import numpy as np
```

```python
arr = np.arange(1 , 25)
print(arr)
print(arr.shape)
print(arr.dtype)
```

```python
reshaped_array = arr.reshape((4 , 6))
print(reshaped_array)
print(reshaped_array.shape)
```

```python
#Reshaping A 2D Array Into 3D Array With 2 Layers
r_2 = reshaped_array.reshape((2 , 3 , 4))
print(r_2)
print(r_2.shape)
```

```python
flattened = r_2.flatten()
print(flattened)
print(flattened.shape)
```

```python
row_max = np.max(reshaped_array , axis = 1)
print(row_max)
```

```python
col_min = np.min(reshaped_array , axis = 0)
print(col_min)
```

```python
sorted_arr = np.sort(reshaped_array)
print(sorted_arr)
```

```python
sort_col = np.sort(reshaped_array , axis = 0)
print(sort_col)
```

```python
arr1 = np.array([40, 10, 30, 20])
sort_arg = np.argsort(arr1)#This Particular function in numpy returns the indices that makes an array sorted
print(sort_arg)
print(arr1[sort_arg])#This Is The Sorted Arr
```
