```python
import numpy as np
```

```python
#Data Creation
np.random.seed(42)
marks = np.random.randint(0, 101, size=20)
print(marks)
```

```python
#Basic Statstics
average_marks = np.mean(marks)
highest_mark = np.max(marks)
lowest_mark = np.min(marks)
print(average_marks)
print(highest_mark)
print(lowest_mark)
```

```python
#Avg Count

mask = marks>average_marks
above_avg_count = np.sum(mask)
print(above_avg_count)


```

```python
#Fail Rule Application
marks_set = np.array(marks)
mask2 = marks<40
marks_set[mask2] = 0
print(marks_set)
```

```python
#Z - Score Normalization
#Marks Normalization
avg = np.mean(marks)
std = np.std(marks)
normalized_vector = (marks - avg)/std
print(normalized_vector)
```

```python
# Boolean masking is better because NumPy executes it in optimized C code,
# avoiding slow Python-level loops and per-element overhead.

# With 2,000,000 students, Python loops become extremely slow and memory-inefficient,
# while vectorized NumPy operations still run efficiently at scale.

# Operations like counting, filtering, replacing values, and normalization would be
# very slow in pure Python loops due to repeated interpreter overhead.
```

```python
#Data Creation
np.random.seed(42)
scores = np.random.randint(1 , 101 , size = (5 , 30))
print(scores)
print(scores.shape)

```

```python
avg = np.mean(scores , axis = 1)#Per section
highest_marks = np.max(scores , axis = 1)
lowest_marks = np.min(scores , axis = 1)
print(avg)
print(highest_marks)
print(lowest_marks)
```

```python
#In each section we have to count students which are greater than avg and then store them in a single array
above_avg_marks = scores > avg[:,np.newaxis]
above_avg_count = np.sum(above_avg_marks , axis = 1)
print(above_avg_count)
```

```python
#Fail Rule 
marks_s = np.array(scores)
mask = scores < 35
marks_s[mask] = 0
print(marks_s)
```

```python
#Z - Score Normalisation
avg = np.mean(scores , axis = 1)
avg = avg.reshape((5 , 1))
deviation = np.std(scores , axis = 1)
deviation= deviation.reshape((5 , 1))
z_scores = (scores - avg) / deviation
print(z_scores)
```
