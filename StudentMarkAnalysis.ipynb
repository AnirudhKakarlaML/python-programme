```python
import numpy as np
```

```python
marks = np.array([
    [78, 85, 90, 88, 92],
    [45, 55, 60, 58, 62],
    [88, 92, 94, 91, 89],
    [70, 75, 80, 72, 68],
    [95, 98, 99, 97, 96],
    [60, 65, 58, 62, 64],
    [82, 79, 85, 88, 84],
    [50, 52, 48, 55, 53]
])
```

```python
#Basic Analytics
total_marks = np.sum(marks , axis = 1)
print(total_marks)
average_marks = np.mean(marks , axis = 1)
print(average_marks)
highest_score = np.argmax(total_marks)
print(highest_score)
lowest_score = np.argmin(total_marks)
print(lowest_score)
```

```python
#Subject Analysis
avg_sub = np.mean(marks , axis = 0)
print(avg_sub)
high_avg = np.argmax(avg_sub)
print(high_avg)
low_avg = np.argmin(avg_sub)
print(low_avg)
```

```python
#Classification
r1 = average_marks>=85
r2_1 = average_marks>=70
r2_2 = average_marks<85
r2_comb = r2_1 & r2_2
r3_1 = average_marks>=50
r3_2 = average_marks<70
r3_comb = r3_1 & r3_2
r4= average_marks < 50
conditions = [r1 , r2_comb , r3_comb , r4]
choices=['Distinction' , 'FirstClass' , 'Pass' , 'Fail']
result = np.select(conditions , choices , default = 'Unknown')
print(result)
```
