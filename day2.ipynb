```python
import numpy as np
```

```python
marks_data = np.array([
 [101, 78, 85, 90],
 [102, 88, 79, 92],
 [103, 65, 70, 60],
 [104, 90, 95, 93],
 [105, 72, 68, 75]
]
)
```

```python
#Average Marks Per Student
avg_marks = np.mean(marks_data[:,1:] ,axis = 1)#Avg Marks Per Student
print(avg_marks)
print(avg_marks.shape)
```

```python
#Avg Marks Per Subject
avg_sub = np.mean(marks_data[:,1:], axis = 0)#Avg Per Subject
print(avg_sub)
print(avg_sub.shape)
```

```python
#Total Marks
total_marks = np.sum(marks_data[:,1:] , axis = 1)
print(total_marks)
```

```python
#Index Of Topper
index_topper =np.argmax(total_marks)
print(index_topper)
```

```python
#Extracting Roll Number And Topper Info
roll_no = marks_data[index_topper , 0]
print(roll_no)
Marks_topper = total_marks[index_topper]
print(Marks_topper)
```

```python
#Subject With Lowest Average
lowest_avg = np.argmin(avg_sub)
subjects = np.array(["Maths", "Physics", "Chemistry"])
print(subjects[lowest_avg])
```

```python
#Extracting Marks Coloumns
marks_col = marks_data[:,1:]
print(marks_col , marks_col.shape)
```
