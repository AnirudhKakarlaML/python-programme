```python
import numpy as np
```

```python
#col[0] = Experience(Years)
#col[1] = Projects Completed
#col[2] = Performance Score
#col[3] = salary(in Thousands)

employees = np.array([
    [2, 5, 65, 40],
    [5, 12, 78, 60],
    [1, 3, 50, 35],
    [8, 20, 90, 85],
    [3, 7, 72, 50],
    [10, 25, 95, 100],
    [4, 9, 68, 55],
    [6, 15, 80, 75],
    [2, 4, 60, 42],
    [7, 18, 88, 82]
])
```

```python
#BASIC INSPECTION
size = employees.shape
col_mean = np.mean(employees , axis = 0)
col_min = np.min(employees , axis = 0)
col_max = np.max(employees , axis = 0)
print(size)
print(col_mean)
print(col_min)
print(col_max)
```

```python
#Performance Score Analysis
mask = employees[:,2]>=85
top_performers_count= np.sum(mask)
print(top_performers_count)
#Printing All Rows Of These Employees
print(employees[mask])
```

```python
#UnderPAid Performers
salary_mask = employees[:,3]<50
performance_mask  = employees[:,2]>=75
combined = salary_mask & performance_mask
print(employees[combined])
print(np.sum(combined))
```

```python
#Performance-Tier-Classification(New)
Excellent_Performers = employees[:,2]>=90
Good_Performers = employees[:,2]>=75
good_2 = employees[:,2]<90
comb_good = Good_Performers & good_2
avg_performers = employees[:,2]<75
count_excellent = np.sum(Excellent_Performers)
count_good = np.sum(comb_good)
count_avg = np.sum(avg_performers)
print(count_excellent)
print(count_good)
print(count_avg)
```

```python
#Productivity Index
productivity_index = employees[:,1]/employees[:,0]
print(productivity_index)
```

```python
#Adding Productivity Index Coloumn To The Original Data Set
p_2 = productivity_index.reshape(-1 , 1)
employees_new = np.hstack((employees , p_2))
print(employees_new)
```

```python
#Normalization
e_min = np.min(employees_new , axis = 0)
e_max = np.max(employees_new , axis = 0)
Normalized_arr = (employees_new - e_min)/(e_max - e_min)
print(Normalized_arr)
```
