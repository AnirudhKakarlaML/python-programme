```python
import numpy as np
```

```python
arr = np.array([10 , 20 , 30 , 40 , 50 , 60])
print(arr)
print(arr.dtype)
print(arr.shape)
```

```python
arr = np.zeros((3, 4))
print(arr)
print(arr.shape)
```

```python
arr = np.ones((4 , 3))
print(arr)
print(arr.shape)
```

```python
arr = np.arange(10 , 51 , 5)
print(arr)
print(arr.shape)
```

```python
arr = np.linspace(0 , 1 , 10)
print(arr)
print(arr.shape)
```

```python
np.random.seed(42)
arr = np.random.randint(1 , 101 , size = (4 , 4))
print(arr)
print(arr.shape)
```

```python
arr = np.random.rand(3, 3)

print(arr)
print(arr.shape)
```

```python
arr = np.arange(1, 26).reshape(5,5)
print(arr[2])
```

```python
arr = np.arange(1, 26).reshape(5,5)
print(arr[: , 1:2])
print(arr[0:3 , 0:3])
```

```python
arr = np.arange(1, 26).reshape(5,5)
arr[3] = 0
print(arr)
```
