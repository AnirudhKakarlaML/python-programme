```python
import numpy as np
```

```python
np.random.seed(42)
arr = np.random.randint(1 , 101 , size = (6,4))
print(arr)
```

```python
col_mean = np.mean(arr ,  axis = 0)
print(col_mean)
```

```python
new_arr = arr - col_mean
print(new_arr)
```

```python
row_sum = np.sum(arr ,  axis = 1)
print(row_sum)
```

```python
row_average = np.mean(arr , axis = 1)
print(row_average)
```

```python
mask = arr > 70
print(mask)
```

```python
count = np.sum(mask)
print(count)
```

```python
avg_arr = np.mean(arr)
arr[mask] = avg_arr
print(arr)
```

```python
flattened = arr.flatten()
print(flattened)
```

```python
sort_fla = np.argsort(flattened)
descending = sort_fla[-1::-1]
print(flattened[descending])
```

```python
f_2 = np.sort(flattened)
print(f_2[-1::-1])
```

```python
descending = np.sort(flattened)[::-1]

top5 = descending[:5]
print(top5)
bot5 = descending[-5:]
print(bot5)
```

```python
unq = np.unique(f_2)
print(unq)
```
