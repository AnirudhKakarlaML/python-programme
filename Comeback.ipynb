```python
#Reshape
```

```python
arr = np.array([1,2,3,4,5,6])
reshaped = arr.reshape((2, 3 ))
print(reshaped)
reshaped2 = reshaped.reshape((3 , 2))
print(reshaped2)
```

```python
#Ravel - Any Array Into 1D Array
#Return View
```

```python
Ravel = reshaped2.ravel()
print((Ravel))
Ravel[0] = 1001
print(reshaped2)
```

```python
#Flatten - Turns Into 1D array
#Just A Copy
```

```python
faltten = Ravel.flatten()
print(faltten)
```

```python
faltten[1] = 0
print(faltten)
print(Ravel)
```

```python
#Indexing And Slicing
```

```python
a = [1 , 2 , 4 ,5 ]
print(a[-1 : -4 : -1])
print(a[::2])
```

```python
arr = np.array([1 , 2 , 3 , 4 , 5])
#indexes - 0 , 2, 4
print(arr[::2])
```

```python
#Fancy Indexing
arr = np.array([10 , 20 , 30 , 40 , 50 , 60])
idx=[0 , 1 , 3]
print(arr[idx])
```

```python
#View vs copy
```

```python
arr = np.array([1 , 2 , 3 , 4 , 5])
view = arr[1:3]
print(view)
view[0] = 200
print(view)
```

```python
copy = arr[1:3].copy()
copy[0] = 2
print(copy)
print(arr)
```

```python
arr = np.array([[1 , 2] , [ 3  , 4]])
arr2 = arr.transpose()
print(arr2)
```

```python
#Swap Axes
```

```python
arr = np.array([[[1 , 2] , [ 3  , 4]]])
print(arr.shape)
swap = np.swapaxes(arr , 0 , 1)
print(swap.shape)
```

```python
#Concatenate
a = np.array([1 , 2 , 3])
b = np.array([1 , 2 , 3])
c = np.concatenate((a , b))
print(c)
```

```python
#Vstack and Hstack
arr1 = np.array([[1 , 2] , [3 , 4]])
arr2 = np.array([[5 , 6] , [7 , 8]])
vstack = np.vstack((arr1 , arr2))
print(vstack)
hstack = np.hstack((arr1 , arr2))
print(hstack)
```

```python
#Repeat And Tile
arr = np.array([1 , 2 , 4])
arr2 = np.repeat(arr , 2)
arr3 = np.tile(arr , 2)
print(arr2)
print(arr3)
```

```python
arr = np.array([1 , 2 , 3 , 4 ,5 , 6])
mask= (arr>3) & (arr<6)
print(arr[mask])
```
