```python
import numpy as np
```

```python
np.random.seed(42)#For Reproducability
arr = np.random.randint(10 , 101 , size = (10 , 5))
print(arr)
print(arr.shape)
```

```python
converted = arr.astype(dtype = 'float64')
print(converted.dtype)
```

```python
import numpy as np

# Assuming converted is already created
converted[2, 3] = np.nan
converted[7, 1] = np.nan
converted[5, 4] = np.nan

print(converted)

# Count total NaNs
print("Total NaNs:", np.isnan(converted).sum())
```

```python
col_mean = np.nanmean(converted , axis = 0)
print(col_mean)
mask = converted[2: , :]
mask = col_mean
print(mask)
```
