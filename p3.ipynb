```python
import numpy as np
```

```python
np.random.seed(42)
arr = np.random.randint(-50 , 51 , size = (5 , 3))
print(arr)
```

```python
col_mean = np.mean(arr , axis = 0)
print(col_mean)
print(col_mean.shape)
```

```python
centered_arr = arr-col_mean
print(centered_arr)
print(centered_arr.shape)
```

```python
mask = arr>0
positive_count = np.sum(mask)
print(positive_count)
```

```python
mask = arr < 0
negative_count = np.sum(mask)
print(negative_count)
```

```python
arr[arr < 0] = 0
print(arr)
```

```python
scaled_arr = arr*2#Multiplication By A Scalar
print(scaled_arr)
```

```python
add_arr = np.array([10 , 20 , 30])
broadcasted_arr  = scaled_arr+add_arr
print(broadcasted_arr)
```

```python
#Z-score-Normalisation
mean_bc = np.mean(broadcasted_arr)
dev_bc = np.std(broadcasted_arr)
normalized_arr = (broadcasted_arr - mean_bc)/dev_bc
print(normalized_arr)
```

```python
min_arr = np.min(broadcasted_arr)
max_arr = np.max(broadcasted_arr)
normalized_arr = (broadcasted_arr-min_arr)/(max_arr - min_arr)
print(normalized_arr)
```

```python
max_narr = np.max(normalized_arr)
print(max_narr)
```

```python
min_narr = np.min(normalized_arr)
print(min_narr)
```

```python
print(arr.shape)
print(col_mean.shape)
print(normalized_arr.shape)
```
