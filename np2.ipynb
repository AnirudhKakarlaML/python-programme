```python


arr = np.array([1, 2, 3, 4, 5])

# Case 1: WITHOUT copy
x = arr[1:4]
x[0] = 99

print("Without copy:")
print("x =", x)
print("arr =", arr)

# Case 2: WITH copy
arr = np.array([1, 2, 3, 4, 5])
y = arr[1:4].copy()
y[0] = 99

print("\nWith copy:")
print("y =", y)
print("arr =", arr)
```
