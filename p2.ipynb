```python
import numpy as np
```

```python
#ðŸ”¹ Problem 1: Create a random 4Ã—4 NumPy array
arr = np.random.randint(1 , 101 , size = (4 ,4))
print(arr)
```

```python
#Task:Extract First Row 
first_row = arr[0]
print(first_row)
```

```python
#Extracting Last Col
last_col = arr[:,-1]
print(last_col)
```

```python
center = arr[1:3 , 1:3]#Slicing The Center Part As (2,2) matrix
print(center)
```

```python
#Replace 2nd row with zeros
arr[1] = 0 
print(arr)
```

```python
#Mean Of Each Row
mean_row= np.mean(arr , axis = 1)
print(mean_row)
```

```python
#Coloumn Wise Sum
col_sum = np.sum(arr , axis = 0)
print(col_sum)
```

```python
#max value present in array
max_val = np.max(arr)
print(max_val)
```

```python
#Min Value In An Array
min_val = np.min(arr)
print(min_val)
```

```python
mask = arr>60#Vals Greater Than 60-->Count
count = np.sum(mask)
print(count)
```

```python
mask = arr[arr>60]#Values In Array That Are Greater Than 60
print(mask)
```

```python
arr[arr < 30] = 30#Replace Vals Which Are Less Than 30 with 30
print(arr)
```

```python
#Flattenment Of The Array
flattened_arr = arr.flatten()
print(flattened_arr)
```

```python
sort_arr = np.sort(flattened_arr)#Sorting
print(sort_arr)
```
